```python
import pandas as pd

input_df = pd.read_csv("query4_results.csv")
input_df.head()
```

```python
input_df.count()
```

```python
input_v2 = input_df.dropna()
input_v2.count()
```

```python
subset_df = pd.DataFrame()

for category in input_v2['business_category'].unique():
    # Filter the DataFrame based on the current category and take the first 20 rows
    category_df = input_v2[input_v2['business_category'] == category].head(20)
    
    # Append the resulting DataFrame to subset_df
    subset_df = pd.concat([subset_df, category_df], ignore_index=True)
```

```python
subset_df.head()
```

```python
subset_df['business_category'].value_counts()
```

```python
import json

# convert into format for fine_tuning api
df = subset_df.rename(columns={'domain_name': 'input_text', 'business_category': 'output_text'})
```

```python
# Create a string with all industry categories
industry_list = pd.read_csv("LABEL_824_DF.csv")["business_category"].tolist()
industries_str = ", ".join(industry_list)
industries_str = "[" + industries_str + "]"
industries_str
```

```python
# include instructions into input text
subset_df['formatted_string'] = subset_df['input_text'].apply(
    lambda domain: f"Classify the following domain name into one of the following industries: {industries_str} Domain name: {domain}"
)
```

```python
pd.set_option('display.max_colwidth', None)
subset_df["formatted_string"].head()
```

```python
subset_df["input_text"] = subset_df["formatted_string"]
subset_df = subset_df.drop(columns=['formatted_string'])
subset_df.head()
```

```python
subset_df.count()
```

```python
# save as jsonl
with open('output_v5.jsonl', 'w') as f:
    for record in subset_df.to_dict(orient='records'):
        f.write(f"{json.dumps(record)}\n")
```

```python
# verify
temp_df = pd.read_json("output_v5.jsonl", lines=True)

temp_df.count()
```

```python
temp_df['output_text'].value_counts()
```

```python
# get evaluation dataset
evl_df = pd.DataFrame()

for category in temp_df['output_text'].unique():
    # Filter the DataFrame based on the current category and take the last 2 rows
    category_df = temp_df[temp_df['output_text'] == category].tail(2)
    
    # Append the resulting DataFrame to subset_df
    evl_df = pd.concat([evl_df, category_df], ignore_index=True)
```

```python
evl_df.head()
```

```python
evl_df.count()
```

```python
evl_df = evl_df.rename(columns={'input_text': 'prompt'})
with open('evaluation_v2.jsonl', 'w') as f:
    for record in evl_df.to_dict(orient='records'):
        f.write(f"{json.dumps(record)}\n")
```
